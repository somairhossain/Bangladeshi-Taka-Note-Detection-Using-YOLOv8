# Train Test Split from makesence.ai labels

      

```python
'''
MyDrive/
 └── yolo_dataset/
      ├── images/
      │    ├── train/
      │    ├── val/
      │    └── test/
      ├── labels/
      │    ├── train/
      │    ├── val/
      │    └── test/
      └── data.yaml
      '''
```

```python
import os
import random
import shutil

# Paths /content/drive/MyDrive/
image_dir = '/content/drive/MyDrive/dataset/images/all'
label_dir = '/content/drive/MyDrive/dataset/labels/all'
```

```python


output_dirs = ['train', 'val', 'test']

for d in output_dirs:
    os.makedirs(f'/content/drive/MyDrive/dataset/images/{d}', exist_ok=True)
    os.makedirs(f'/content/drive/MyDrive/dataset/labels/{d}', exist_ok=True)

# Get all images
images = [f for f in os.listdir(image_dir) if f.endswith(('.jpg', '.png', '.webp'))]
random.shuffle(images)

# Split ratios
train_ratio = 0.7
val_ratio = 0.2

train_split = int(len(images) * train_ratio)
val_split = int(len(images) * (train_ratio + val_ratio))

train_files = images[:train_split]
val_files = images[train_split:val_split]
test_files = images[val_split:]

def move_files(file_list, split):
    for file in file_list:
        img_src = os.path.join(image_dir, file)
        lbl_src = os.path.join(label_dir, file.replace('.jpg', '.txt').replace('.png', '.txt').replace('.webp', 'txt'))
#/content/drive/MyDrive/
        img_dst = f'/content/drive/MyDrive/dataset/images/{split}/{file}'
        lbl_dst = f'/content/drive/MyDrive/dataset/labels/{split}/{file.replace(".jpg", ".txt").replace(".png", ".txt").replace('.webp', 'txt')}'

        shutil.copy(img_src, img_dst)
        if os.path.exists(lbl_src):
            shutil.copy(lbl_src, lbl_dst)

move_files(train_files, 'train')
move_files(val_files, 'val')
move_files(test_files, 'test')

print("✅ Dataset split completed!")
```
